Plot the CNN loss extracted in `get_cnn_loss.ipynb`

```python
import pandas as pd
from matplotlib import pyplot as plt
```

Load in dataframe

```python
df = pd.read_pickle('loss_data/cnn_loss_20260122-152652_SO_JET.pkl')
df
```

Plot the loss

```python
plt.rcParams.update({'font.size': 14})

fig = plt.figure()
plt.plot(df['step'], df['train_epoch_loss_train'], label='Training Loss')
plt.plot(df['step'], df['train_epoch_loss_val'], label='Validation Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.ylim(0, 1.2)
plt.grid()

plt.savefig(f"../DINO/Figures/cnn_loss_20260122-152652.png", dpi=300, bbox_inches='tight')
```
