Load in data files and combine into one for ML training

Speeds up the preprocessing step

```python
import xarray as xr
```

```python
domain = 'SO_JET'

directory = f"/gws/nopw/j04/ai4pex/twilder/NEMO_data/DINO/EXP16/features/{domain}/coarsened_data/"

fns = [f"MINT_1d_*sa_c_{domain}.nc",
       f"MINT_1d_*vor_c_{domain}.nc",
       f"MINT_1d_*eke_c_{domain}.nc",
       f"MINT_1d_*mke_c_{domain}.nc"]
```

```python
fns[0]
```

```python
for i in range(len(fns)):
    ds = xr.open_mfdataset(directory + fns[i],
                           concat_dim='t', 
                           combine='nested')

    fn = fns[i].replace('*', '0061-0072_')
    
    ds.to_netcdf(directory + fn)
```
