Get weights for regridding DINO EXP16 to DINO EXP4

```python
import numpy as np
import xarray as xr
import xesmf as xe
```

```python
directory = '/gws/nopw/j04/ai4pex/twilder/NEMO_data/DINO/EXP16/features/SO_JET/'
```

Load in target grid from get_mask_bounds_for_conservative_regridding.ipynb

```python
array = np.load(directory + 'SO_JET_target_grid_025.npy', allow_pickle=True)
target_grid = array.item()
```

```python
target_grid
```

Load in EXP16 input grid

```python
array = np.load(directory + 'SO_JET_input_grid_16.npy', allow_pickle=True)
input_grid = array.item()
```

Now, compute regridder weights

```python
regridder = xe.Regridder(input_grid, target_grid, "conservative")
```

Now save regridder for future use

```python
fn = regridder.to_netcdf(directory + 'weights_SO_JET.nc')
```
